```python
# https://sbme-tutorials.github.io/2018/cv/notes/6_week6.html

from skimage import data
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 18
```

```python
coffee =data.coffee()[:,:,1]
```

```python
plt.imshow(coffee)
plt.axis("off")
plt.show()
```

```python

def convolution_filter(gray_img, kernel):
    # temporary operation
    # gray_img = cv2.resize(gray_img, (10, 10))

    kernel_size = len(kernel)

    row = gray_img.shape[0] - kernel_size + 1
    col = gray_img.shape[1] - kernel_size + 1

    result = np.zeros(shape=(row, col))

    for i in range(row):
        for j in range(col):
            current = gray_img[i:i+kernel_size, j:j+kernel_size]
            multiplication = np.abs(sum(sum(current * kernel)))
            result[i, j] = multiplication

    return result
```

```python
k1 =np.array([[-1,-2,-1],
              [0,0,0],
              [1,2,1]])
```

```python
plt.imshow(convolution_filter(coffee, k1))
```

```python
k2 =np.array([[1,1,1],
              [1,1,1],
              [1,1,1]])
```

```python
plt.imshow(convolution_filter(coffee, k2))
```

```python
k3 =(1/16) *np.array([[1,2,1],
              [2,4,2],
              [1,2,1]])
```

```python
plt.imshow(convolution_filter(coffee, k3))
```

# Edge detection

```python
k4 =np.array([[-1,0,1],
              [-1,0,1],
              [-1,0,1]])
```

```python
edgelr =convolution_filter(coffee, k4)

plt.imshow(convolution_filter(coffee, k4),cmap="gray")
```

```python
k5 =np.array([[-1,-1,-1],
              [0,0,0],
              [1,1,1]])
```

```python
edgeupdown =convolution_filter(coffee, k5)

plt.imshow(convolution_filter(coffee, k5),cmap="gray")
```

```python
def binarize( gray_image , threshold ):
    threshold = np.max( gray_image ) * threshold
    return 1 * ( gray_image > threshold )
```

```python
binary_coffee = binarize(coffee,0.3)

plt.imshow(binary_coffee)
```
